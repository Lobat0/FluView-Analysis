```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
ili.drop(columns=['AGE 25-64'], inplace=True)
```

```python
ili.drop(columns=['REGION', 'REGION TYPE'], inplace=True)
```

```python
ili.head()
```

```python
ili['DATE'] = ili['YEAR'].astype(str) + '-' + ili['WEEK'].astype(str)
ili['DATE'] = pd.to_datetime(ili['DATE'] + '-1', format='%Y-%W-%w')
```

```python
ili.head()
```

```python
ili_limpo = ili.copy()
```

```python
ili_limpo['MONTH'] = ili_limpo['DATE'].dt.month
```

```python
ili_limpo.head()
```

```python
ili_limpo['SEASON'] = pd.cut(
    ili_limpo['MONTH'], 
    bins=[0, 3, 6, 9, 12], 
    labels=['Inverno', 'Primavera', 'Verao', 'Outono'],
    include_lowest=True
)
```

```python
ili_limpo.tail()
```

```python
ili_limpo['YEAR_WEEK'] = ili_limpo['YEAR'].astype(str) + '-' + ili_limpo['WEEK'].astype(str)
```

```python
ili_limpo.head()
```

```python
age_cols = ['AGE 0-4', 'AGE 5-24', 'AGE 25-49', 'AGE 50-64', 'AGE 65']
for col in age_cols:
    if col in ili_limpo.columns:
        ili_limpo[f'{col}_%'] = ili_limpo[col] / ili_limpo['ILITOTAL'] * 100
```

```python
ili_limpo.head()
```

```python
ili_limpo['WEEKLY_CHANGE'] = ili_limpo['% WEIGHTED ILI'].diff() #mostra reduo ou aumento nos casos. + teve aumento. - diminuiu casos
```

```python
ili_limpo.head()
```

```python
ili_limpo['ILI_SEVERITY'] = pd.cut(
    ili_limpo['% WEIGHTED ILI'], 
    bins=[0, 1, 2, 3, 5, 10], 
    labels=['Muito Baixo', 'Baixo', 'Moderado', 'Alto', 'Muito Alto'],
    include_lowest=True
)
```
