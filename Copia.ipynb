```python
from datetime import datetime
def get_date(year, week):
    return datetime.strptime(f'{int(year)}-{int(week)}-1', "%Y-%W-%w")
```

```python
clinico['DATE'] = clinico.apply(lambda x: get_date(x['YEAR'], x['WEEK']), axis=1)
```
